# Imports

```python
import pennylane as qml
import matplotlib.pyplot as plt
import numpy as np
```

# Device

```python
dev = qml.device("default.qubit", wires=3)
```

# Circuit

```python
@qml.qnode(dev)
def circuit():
    qml.H(wires=0)
    qml.CNOT(wires=[0, 1])
    qml.CNOT(wires=[1, 2])

    return qml.counts()
```

## Draw the circuit

```python
qml.drawer.use_style("black_white")
qml.draw_mpl(circuit)()
```

# Execute the circuit

```python
counts = circuit(shots=1000)
print(counts)
```

```python
# Plot the counts as a histogram
plt.bar(counts.keys(), counts.values())
plt.show()
```
